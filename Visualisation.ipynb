```r
library(ggplot2)
library(dplyr)

# Create a hypothetical dataset
geology_data <- data.frame(
  rock_type = c('Magmatique', 'Métamorphique', 'Sédimentaire', 'Magmatique', 'Sédimentaire', 'Métamorphique',
                'Magmatique', 'Sédimentaire', 'Métamorphique', 'Magmatique', 'Sédimentaire', 'Métamorphique',
                'Magmatique', 'Sédimentaire', 'Métamorphique'),
  age_million_years = c(100, 250, 300, 150, 200, 275,
                        90, 310, 270, 120, 220, 260,
                        110, 190, 280),
  latitude = c(34.05, 36.16, 35.68, 34.78, 35.04, 36.12,
               33.98, 35.88, 36.50, 34.62, 35.20, 36.10,
               34.45, 35.12, 36.30),
  longitude = c(-118.24, -115.15, -117.23, -119.41, -118.29, -116.39,
                -118.35, -117.00, -116.80, -119.10, -118.15, -116.75,
                -118.50, -118.00, -116.90)
)

# Create the scatter plot
ggplot(geology_data, aes(x = latitude, y = age_million_years, color = rock_type)) +
  geom_point(size = 5) +
  labs(title = "Geological Rock Samples",
       x = "Latitude",
       y = "Age (Million d'années)",
       color = "Type de roche") +
  theme_minimal()
```
